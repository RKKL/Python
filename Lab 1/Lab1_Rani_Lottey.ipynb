```python
#Code to fix year 2070 when data type is changed.
def fix_date(x):
    if x.year >= 2070:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year,x.month,x.day)
GE['Date'] = GE['Date'].apply(fix_date)
IBM['Date'] = IBM['Date'].apply(fix_date)
ProcterGamble['Date'] = ProcterGamble['Date'].apply(fix_date)
CocaCola['Date'] = CocaCola['Date'].apply(fix_date)
Boeing['Date'] = Boeing['Date'].apply(fix_date)
```
