# Visualizing Decision Trees in Jupyter Notebook with Python and Graphviz

Decision Tree Regressors and Classifiers are being widely used as separate algorithms or as components for more complex models. Visualizing them is crucial in order to correctly understand how certain decisions are being made inside the algorithm, which is always important for business applications.



In this short tutorial, I would like to briefly describe the process of visualizing Decision Tree models from sklearn library. <font color = "red"><b>Note: Graphviz installed and configured is required to run the code below.</b></font>

As a toy dataset, I will be using a well known Iris dataset. Let’s import the main libraries and download the data for the experiment.

```python

import pandas as pd
import numpy as np
from sklearn import datasets

FEATURE_NAMES = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']

iris = datasets.load_iris()
X = pd.DataFrame(iris.data, columns = FEATURE_NAMES)
y = iris.target
```

Now we will just create a simple Decision Tree Classifier and fit it on the full dataset.

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X,y)
```

Finally, the interesting steps are coming. We export our fitted decision tree as a <b>.dot</b> file, which is the standard extension for graphviz files. The <b>tree.dot</b> file will be saved in the same directory as your Jupyter Notebook script. Don’t forget to include the <i>feature_names</i> parameter, which indicates the feature names, that will be used when displaying the tree.

```python

from sklearn.tree import export_graphviz
export_graphviz(model, 'tree.dot', feature_names = FEATURE_NAMES)
```

Now, by running the following command we will convert the <b>.dot</b> file to <b>.png</b> file. This will work only in Jupyter Notebook, as the <b>“!”</b> symbol indicates that the command will be performed directly in the console.

```python
! dot -Tpng tree.dot -o tree.png
```

After this manipulation, the <b>tree.png</b> file will appear in the same folder. Now we can easily display it, using well-known libraries.

<font color = "red"><b>Note from Sara</b>: Before executing the below code, the <b>opencv</b> library must also be installed. To do this, I opened Anaconda Navigator and Selected "Environments" -> Select "Not Installed" from the dropdown, and search for "cv". To be sure, I installed the following three libraries:</font>

    libopencv
    opencv

```python
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
img = cv2.imread('tree.png')
plt.figure(figsize = (20, 20))
plt.imshow(img)
```
